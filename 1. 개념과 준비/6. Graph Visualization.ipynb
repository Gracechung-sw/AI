## Load Packages

### matplotlib를 이용한 시각화

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### 점선 그래프 그리기

```python
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show() #그래프를 보여달라. 
```

### 여러 그래프 그릴 준비 하기 (다른 그래프와 비교하면서 볼 때 유용하게 사용할 수 있다. )

  subplot(): 여러 그래프를 그리기 전에 그 안에 또 다른 그래프를 그리겠다고 하는 것, 
  subplot(행의 개수, 열의 개수, 그 중 몇 번째 것을 그리겠다. )
                                                  #matplot은 0부터가 아니라 1부터 시작한다. 
 

```python
plt.subplot(3, 2, 1)
plt.subplot(3, 2, 2)
plt.subplot(3, 2, 3)
plt.subplot(3, 2, 4)
plt.subplot(3, 2, 5)
plt.subplot(3, 2, 6)
plt.show()
```

## Multi Graph 그리기

```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2, 2, 1)
plt.plot(data)

plt.subplot(2, 2, 2)
plt.hist(hist_data, bins = 20) #히스토그램

plt.subplot(2, 2, 3)
plt.scatter(scat_data, np.arange(30)+3) #scatter 그래프
# plt.(그래프 형식)(x축, y축)
```

## 그래프 선 옵션
: 그래프를 그릴 때 표시 되는 색이나 마커 패털은 바꾸는 것 확인 
- 색상 : b(파란), g(초록), r(빨강), c(청록), y(노란), k(검정), w(흰색)
- 마커 : o(원), v(역삼각형), ^(삼각형), s(네모), +(플러스), .(점)

```python
plt.plot(data)
plt.show()
```

```python
plt.plot(data, 'g^') #뒤에다가 string으로 색상과 마커를 주면 그래프디자인을 할 수도 있다. 
```

## 그래프 사이즈 조절

- plt.figure안에 figsize를 이용하여 가로, 세로 길이 조절 가능(inch 단위)

```python
plt.figure(figsize=(1, 1))
plt.plot(data, 'y+')
plt.show()
```

- 여러 그래프의 사이즈 조절

```python
plt.figure(figsize=(3, 3))
plt.subplot(2, 2, 1)
plt.plot(data)

plt.subplot(2, 2, 2)
plt.hist(hist_data, bins = 20) #히스토그램

plt.subplot(2, 2, 3)
plt.scatter(scat_data, np.arange(30)+3) #scatter 그래프
# plt.(그래프 형식)(x축, y축)
```

## 여러 그래프를 겹치기 + legend 달기
: subplot을 주지 않으면 따로 그려지는게 아니라 하나에 합쳐서 그려진다. 

```python
data = np.random.randn(30).cumsum()

plt.plot(data, 'k--')
plt.plot(data, 'y+', drawstyle = 'steps-post')
plt.show()
```

```python
data = np.random.randn(30).cumsum()

plt.plot(data, 'k--', label = 'Default')
plt.plot(data, 'y+', drawstyle = 'steps-post', label = 'steps-post')
plt.legend() #이름을 보여라는 것
plt.show()
```

## 이름 달기

```python
data = np.random.randn(30).cumsum()

plt.plot(data, 'k--', label = 'Default')
plt.plot(data, 'y+', drawstyle = 'steps-post', label = 'steps-post')
plt.title('Random Gragh')
plt.xlabel('Stages')
plt.ylabel('random')
plt.legend() #이름을 보여라는 것
plt.show()
```

## 종합

```python
plt.plot(np.random.randn(1000).cumsum(), 'k', label = 'one')
plt.plot(np.random.randn(1000).cumsum(), 'b-', label = 'two')
plt.plot(np.random.randn(1000).cumsum(), 'r+', label = 'three')

plt.legend()
plt.show()
```
