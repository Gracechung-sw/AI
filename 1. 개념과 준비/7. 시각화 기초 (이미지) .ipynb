# 이미지 시각화

## package 불러오기

```python
import numpy as np
from PIL import Image ##image를 읽어 올때 사용하는 것(앞에 대문자 I !!! 조심)

##시각화 해야 하니까 matplot이 빠질 수 없다. 
import matplotlib.pyplot as plt

%matplotlib inline
```

## 이미지 파일 열기

```python
path = 'image_source/black.jpg'

image_pil = Image.open(path)
image = np.array(image_pil) #이러면 이미지가 array로 담기게 된다. 
```

```python
image.shape ##(세로길이, 가로길이, rgb(색상)) #이미지 정보 확인
```

```python
np.min(image), np.max(image) #이미지를 열기 전에 shape및 min, max를 통해서 이미지의 range확인 필요
```

## 그래프로 시각화 하기

```python
plt.hist(image.ravel(), 256, [0, 256]) ##ravel : 쫙 펼치는 것
plt.show()
```

## 그림 나타내기

```python
plt.imshow(image)
plt.show()
```

## 이미지 흑백으로 열기

```python
path
```

```python
image_pil = Image.open(path).convert("L")
image_bw = np.array(image_pil)
```

```python
image_bw.shape #흑백으로 변했기 때문에 rgb값이 없다. 
```

```python
plt.imshow(image_bw)
plt.show()
```

## 다른 색상으로 c(olor)map표현하기

### gray scale

```python
plt.imshow(image_bw, 'gray')
plt.show()
```

### Red, Blue 로 표현 (RdBu)

```python
plt.imshow(image_bw, 'RdBu')
```

### jet : 밝을 수록 빨간색, 어두울 수록 파란색

```python
plt.imshow(image_bw, 'jet')
```

### color bar 추가하기

```python
plt.imshow(image_bw, 'jet')
plt.colorbar()
```

### 두번째 이미지 열기

```python
second_path = 'image_source/brown.jpg'

second_pil = Image.open(second_path)
second_image = np.array(second_pil)
```

```python
plt.imshow(second_image)
plt.show()
```

## 두 번째 이미지를 첫 번째 이미지 모양에 맞추기

```python
second_image.shape
```

```python
##근데 앞의 이미지는 (1171, 750, 3) 로, 두번째 이미지인 (783, 745, 3) 와 크기가 안 맞는다. 
image.shape, second_image.shape
```

```python
#첫번째 이미지를 줄여보고자 한다. 
#open cv를 이용!!

import cv2
```

```python
black_image = cv2.resize(image, (745, 783)) #왜그런진 모르겠지만 반대로 줘야 한다. 즉, (783, 745)가 아니라 (745, 783)
black_image.shape
```

```python
black_image.shape, second_image.shape
```

### 이미지 합치기

```python
plt.imshow(black_image)
plt.imshow(second_image, alpha = 0.5) #alpha : 투명도 
plt.show()
```
